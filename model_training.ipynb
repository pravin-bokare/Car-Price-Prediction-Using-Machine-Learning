```python
import pandas as pd
```

```python
df = pd.read_csv('cars24-car-price.csv')
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df.columns
```

```python
df= df[['fuel_type', 'transmission_type', 'engine', 'seats', 'seller_type', 'mileage', 'selling_price']]
```

```python
df.head()
```

```python
fuel_type = {'Diesel':1, 'Petrol':2, 'CNG':3, 'LPG':4, 'Electric':5}
transmission_type = {'Manual':1, 'Automatic':2}
seller_type = {'Dealer':1, 'Individual':2, 'Trustmark Dealer':3}
```

```python
df['fuel_type'] = df['fuel_type'].map(fuel_type)
df['transmission_type'] = df['transmission_type'].map(transmission_type)
df['seller_type'] = df['seller_type'].map(seller_type)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.corr()
```

```python
X = df.drop(['selling_price'], axis=1)
y = df['selling_price']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state=42)
rfr.fit(X_train, y_train)
```

```python
rfr.score(X_test, y_test)
```

```python
import pickle
pickle_out = open('car_pred.pkl', mode='wb')
pickle.dump(rfr, pickle_out)
pickle_out.close()
```
